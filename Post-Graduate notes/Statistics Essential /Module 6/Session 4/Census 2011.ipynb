#### Segmented Analysis
- What percentage of females in the age group 20-24 are illiterate in India, i.e. out of all the females in the age group 20-24, what fraction is illiterate?

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="white")

census_df = pd.read_csv('Census.csv',sep='\t')

print(census_df.shape)
```

```python
female_age_group = census_df[(census_df['Age-group'] == '20-24') & (census_df['Area Name'] == 'INDIA') & (census_df['Total/Rural/Urban/'] == 'Total')].head()
percentageOfIlliterateFemales = 100 * (female_age_group['Illiterate/Females']/female_age_group['Total/Females'])
print(percentageOfIlliterateFemales)
```

- Compare the literacy rates (i.e. the number of literates / total number of population) in each age group and choose the correct option.

```python
census_df.rename(columns={'    Total/Persons': 'Total/Persons'}, inplace=True)
```

```python
census_df['percentage_of_literate_all_age_group'] = 100 * (census_df['Literate/Persons']/census_df['Total/Persons'])
print(census_df[['Age-group','percentage_of_literate_all_age_group']].iloc[9:28,:,])
print("Mean <55", census_df[['percentage_of_literate_all_age_group']].iloc[9:16,:,].mean())
print("Mean >55", census_df[['percentage_of_literate_all_age_group']].iloc[22:28,:,].mean())
```

```python
census_df['percentage_of_literate_female_each_region'] = 100 * (census_df['Literate/Females']/census_df['Total/Females'])
census_df = census_df[(census_df['Age-group'] == 'All ages') & (census_df['Total/Rural/Urban/'] == 'Total')]
census_df1 = census_df[['Area Name','percentage_of_literate_female_each_region']]
census_df1.sort_values('percentage_of_literate_female_each_region', ascending=False)
```

- Across all the states, which state shows the lowest literacy rate in 2011 (total population)?

```python
census_df['percentage_of_illiterate_each_region'] = 100 * (census_df['Illiterate/Persons']/census_df['Total/Persons'])
census_df = census_df[(census_df['Age-group'] == 'All ages') & (census_df['Total/Rural/Urban/'] == 'Total')]
census_df2 = census_df[['Area Name','percentage_of_illiterate_each_region']]
census_df2.sort_values('percentage_of_illiterate_each_region', ascending=False)

plt.figure(figsize=(20, 16))
sns.barplot(y='Area Name', x='percentage_of_illiterate_each_region', data=census_df2)
plt.xlabel('literacy in %')
plt.title("Literacy rate of India 2011")
```
