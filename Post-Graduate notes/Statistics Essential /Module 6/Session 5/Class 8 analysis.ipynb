```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="ticks", color_codes=True)

nas_df = pd.read_csv('nas.csv')

print(nas_df.shape)
```

```python
nas_df.head()
```

```python
nas_df['Siblings'] = nas_df['Siblings'].apply(lambda x : x[0] if x.find("sibling") != -1 else x)
```

```python
nas_df.head()
```

```python
nas_df['Siblings'] = nas_df['Siblings'].apply(lambda x : x[0] if x.find("+") != -1 else x)
```

```python
nas_df['Siblings'] = nas_df['Siblings'].apply(lambda x : 0 if x.find("child") != -1 else x)
```

```python
nas_df['Siblings'] = nas_df['Siblings'].astype(str).astype(int)
```

```python
nas_df.columns
```

```python
nas_df['Mother.edu'] = nas_df[~nas_df['Mother.edu'].isnull()]
```

```python
nas_df.info()
```

```python
#nas_df.sample(100).plot.scatter(x='Siblings', y='Mother.edu')
sns.catplot(x="Mother.edu", y="Siblings", data=nas_df);
```
