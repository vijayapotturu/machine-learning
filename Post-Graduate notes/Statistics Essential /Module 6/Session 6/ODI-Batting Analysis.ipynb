```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="ticks", color_codes=True)

batting_df = pd.read_csv('odi-batting.csv')

print(batting_df.shape)
```

```python
batting_df.head()
```

```python
batting_df_max_100 = batting_df[batting_df['Runs'] > 99]

batting_df_max_100 = batting_df_max_100.groupby('Player')['Runs'].count().reset_index()
sns.swarmplot(x='Player',y='Runs',data=batting_df_max_100.sort_values('Runs',ascending=False).head(5))
fig=plt.gcf()
#plt.set_size_inches(14,8)
plt.ylim(-10,200)
plt.show()
```

```python
batting_df['Year'] = pd.to_datetime(batting_df['MatchDate']).dt.year
batting_df.head()
batting_df.reset_index()
batting_df.reset_index(drop=True)
```

```python
batting_df.info()
```

```python

batting_df_max_100_Indians = batting_df[(batting_df['Runs'] > 99) & (batting_df['Country'] == 'India')]

batting_df_max_100_Indians = batting_df_max_100_Indians.groupby('Year',as_index=False)['Runs'].count().reset_index()
sns.swarmplot(x='Year',y='Runs',data=batting_df_max_100_Indians.sort_values('Runs',ascending=False).head(5))
fig=plt.gcf()
#plt.set_size_inches(14,8)
plt.ylim(-10,200)
plt.show()
```
