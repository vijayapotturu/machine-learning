```python
student_loan_df['Recovery (%)'] = student_loan_df['Recovery (%)'].str.strip('%')
print(student_loan_df.head())
```

```python
student_loan_df['Recovery (%)'] = student_loan_df['Recovery (%)'].astype(float)
student_loan_df['Recovery (%)'] = student_loan_df['Recovery (%)']/100
print(student_loan_df.head())
```

```python
# in the case of default, what would be the value of X, i.e. the money lost by the bank?
student_loan_df['X (Money Lost by Bank in lakh Rs.)'] = student_loan_df['Exposure at Default (in lakh Rs.)'] - student_loan_df['Exposure at Default (in lakh Rs.)']*student_loan_df['Recovery (%)']
```
