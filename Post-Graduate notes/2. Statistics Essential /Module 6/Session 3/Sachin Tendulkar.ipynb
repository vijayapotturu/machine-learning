```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="white")

sachin_df = pd.read_csv('tendulkar_ODI.csv')

sachin_df.head()
print(sachin_df.shape)
```

```python
sachin_df = sachin_df[(sachin_df.Runs != 'DNB')]
sachin_df = sachin_df[(sachin_df.Runs != 'TDNB')]
```

```python
print(sachin_df.shape)
sachin_df.head(10)
sachin_df.info()
```

```python
sachin_df['Runs'] = sachin_df.Runs.apply(lambda x : int(x.strip().replace('*', '')))
print(sachin_df.head())
```

```python
sachin_df.sort_values('Runs', inplace=True, ascending=True)
print(sachin_df.head())
```

```python
# Runs Scored by Tendulkar
# Plot a bar chart showing runs scored on the x-axis and frequency/count on the y-axis.
# In which bucket has he scored runs the most often?

# adjust figure size
plt.figure(figsize=(20, 16))
sachin_df.Runs.value_counts().plot.bar()
#sns.barplot(x='funding_round_type', y='total_no_of_investment', data=master_primary_main_sector_df)
#plt.xscale('log')
plt.ylabel('frequency/count')
plt.title("runs scored")
```

```python
sachin_df = sachin_df[sachin_df['4s'] != '-']

pd.to_numeric(sachin_df['4s'], errors='coerce')
sachin_df.info()
```
